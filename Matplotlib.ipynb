# Matplotlib

```python
import matplotlib.pyplot as plt
from pydataset import data
aq = data('airquality')
ap = data('AirPassengers')
%matplotlib inline
```

```python
plt.plot(ap['time'],ap['AirPassengers'] )
plt.show()
```

```python
plt.hist(aq['Temp'],bins=10)
plt.show()
```

```python
iris = data('iris')
```

```python
iris.shape
```

```python
iris.columns
```

```python
plt.plot(iris['Sepal.Length'], iris['Sepal.Width'])
plt.show()
```

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
plt.boxplot(iris['Sepal.Width'])
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
iris.plot(y='Sepal.Width', kind='box')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
iris.plot(y='Sepal.Width', kind='hist', bins=30)
plt.ylabel('sepal width (cm)')
plt.show()
```
